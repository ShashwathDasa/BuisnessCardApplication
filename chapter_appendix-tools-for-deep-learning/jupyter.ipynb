```python
grouped = df.groupby(["Contract", "Churn Value"]).size().reset_index(name="count")

# total per Contract
grouped["total"] = grouped.groupby("Contract")["count"].transform("sum")

# percentage
grouped["percentage"] = (grouped["count"] / grouped["total"]) * 100

grouped
```

```python
df['tenure_bin'] = pd.cut(df['Tenure Months'], bins=[0, 12, 24, 48, 72])
df.head()
```

```python
df = df[df['Contract'] == 	"Month-to-month"]
grouped = df.groupby(["Contract", "tenure_bin", "Churn Value"]).size().reset_index(name="count")

# total per Contract
grouped["total"] = grouped.groupby("tenure_bin")["count"].transform("sum")

# percentage
grouped["percentage"] = (grouped["count"] / grouped["total"]) * 100

grouped
```
